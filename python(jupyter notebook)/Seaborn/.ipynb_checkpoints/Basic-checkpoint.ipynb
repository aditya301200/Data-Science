# Seaborn library
- It is a library for making statistical graphics in Python. It builds on top of matplotlib and integrated closely with pandas data structures.
- Different categories of plot in seaborn.
- Plots are basically used for visualizing the relationship between variables. Thos variables can be either completely numerical or a catefory like a group, class or division. Seaborn divides the plot into the below categories:
    - **Relational Plots**: This plot is used to understand the relation between two variables.
    - **Categorical plots**: This plot deals with categorical variables and how they can be visualized.
    - **Distribution plots**: This plto is used for examining univariate and bivariate distributions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var1 = [1,2,3,4,5,6,7,8]
var2 = [8,7,6,5,4,3,2,1]
```

```python
plt.plot(var1,var2)
plt.show()
```

```python
sns.lineplot(x = var1,y = var2)
plt.show()
```

```python
df = pd.DataFrame({"var1": var1, "var2":var2})
df
```

```python
sns.lineplot(x="var1", y="var2",data=df)
plt.show()
```

```python
df1 = sns.load_dataset("penguins")
df1
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm",data = df1,
             hue = "sex", palette = "rocket_r",style = "sex", markers= ["o",">"])
plt.show()
```

```python
df2 = sns.load_dataset("penguins").head(20)
df2.describe()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm",data = df2,
             hue = "sex", palette = "rocket_r",style = "sex", markers= ["o",">"])
plt.grid()
plt.title("penguin")
plt.show()
```

### Bar Plot

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm,hue=df1.sex)
plt.show()
```

### Ordering in barplot

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = df1.island, y = df1.bill_length_mm, hue=df1.sex, order = order1)
plt.show()
```

### Hue order changing

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
order2 = ["Female","Male"]
sns.barplot(x = df1.island, y = df1.bill_length_mm, hue=df1.sex, order = order1,hue_order = order2,palette="Accent")
plt.show()
```

### Histogram Plot

```python
sns.displot(df1.flipper_length_mm)
plt.show()
```

#### Bins Parameter

```python
sns.displot(df1.flipper_length_mm, bins = [170,180,190,200,210,220,230,240],kde=True)
plt.show()
```
