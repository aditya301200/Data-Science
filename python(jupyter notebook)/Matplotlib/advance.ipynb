```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
fname = r'IHH.jpg'

# opening image using PIL
image = Image.open(fname).convert('L')

# Mapping image to gray scale
plt.imshow(image, cmap='gray')
plt.show()
```

```python
fname = r'bird.png'

# opening image using PIL
image = Image.open(fname).convert('L')

# Mapping image to gray scale
plt.imshow(image, cmap='gray')
plt.grid()
plt.show()
```

### Check cmap values available to us

```python
fname = r'bird.png'

# opening image using PIL
image = Image.open(fname).convert('L')

# Mapping image to gray scale
plt.imshow(image, cmap='g')
plt.show()
```
