# Matplotlib
- It is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.

## Plots
- Linear Plot
- Scatter Plot
- Bar Plot
- Stem Plot
- Step Plot
- Hist PLot
- Box Plot
- Pie Plot
- Fill_Between_Plot

```python
import matplotlib.pyplot as plt
```

### Linear Plot

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [15,16,13,14]
c = 'r'
plt.plot(x,y,c,marker='o')
plt.show()
```

```python
x = [1,2,3,4]
y = [9,6,8,5]

plt.plot(x,y)
plt.plot(y,marker='o')
plt.plot(x,marker='>')
plt.show()
```

### Bar Plot

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y,color=c)
plt.show()
```

```python
import numpy as np
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y) # horizontal represtation
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.bar(x,y,color='g',width=0.3)
plt.show()
```

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])

plt.barh(x,y,color='g',height=0.1)
plt.show()
```

### Scatter Plot

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
c=['r','magenta','g','orange','b','brown']

plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Numbers")
plt.scatter(x,y,color=c, s = 150)
plt.show()
```
