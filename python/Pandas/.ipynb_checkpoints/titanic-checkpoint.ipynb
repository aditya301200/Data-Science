### Delete column from dataset

```python
df.drop("PassengerId", axis=1) # temporary delete
```

```python
df.drop("Survived", axis = 1, inplace = True) # Permanently deleted
```

```python
df
```

### Delete rows from dataset

```python
df.drop(3) # Temporary deleted
```

```python
df
```

```python
df.drop(3, inplace = True) # permanently deleted
```

```python
df
```

### Set a column as Index of the dataset
- Use "inplace = True" if you want to change permanently

```python
df.set_index("Name") # temporary changed
```

```python
df
```

### Reset index of the dataset

```python
df.reset_index()
```
